```python
import os
import shutil
import random


!pip install tqdm --upgrade

from tqdm.notebook import tqdm
```

```python
train_path_img = '/content/ultralytics/ultralytics/assets/images'
train_path_label = '/content/ultralytics/ultralytics/assets/labels'
test_path = '/content/ultralytics/ultralytics/assets/tets'
```

```python
def train_test_split(path, neg_path=None, split = 0.2):
  print('–ø–æ—á–∞–ª–æ—Å—è')

  files = list(set([name[:4] for name in os.listdir(path)]))

  print(f"–≤ –ø–∞–ø—Ü—ñ —Å—Ç—ñ–∫–æ —Ñ–∞–π–ª—ñ–≤ {len(files)}")
  random.seed(42)
  random.shuffle(files)

  test_size = int(len(files)* split)
  train_size = len(files) - test_size

  os.makedirs(train_path_img, exist_ok= True)
  os.makedirs(train_path_label, exist_ok= True)

  for filex in tqdm(files[:train_size]):
    if filex == 'classes':
      continue
    shutil.copy2(path + filex + '.jpg', f"train_path_img/"+filex + '.jpg')
    shutil.copy2(path + filex + '.jpg', f"train_path_label/"+filex + '.jpg')



    if neg_path:
      neg_images = list(set([name[:4] for name in os.listdir(neg_path)]))
      for filex in tqdm(neg_path):
        shutil.copy2(neg_path+filex+ ".jpg", f"{train_path_img}/" + filex + '.jpg')


    train_test_split('/content/boom')















```
